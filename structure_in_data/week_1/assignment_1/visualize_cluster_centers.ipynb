```python
import pandas as pd
import math

df = pd.read_csv('clusters.csv')
df.head()
```

```python
lats = df.cent_latitude.tolist()
lons = df.cent_longitude.tolist()
sizes = map(lambda x: 5*math.log(x), df.clust_size.tolist())
```

```python
len(lats), len(lons), len(sizes)
```
